# Logistic Regression
1. Logistic Regression is used for Classification only
2. Sigmoid curve is fitted between the classes



```python
from warnings import filterwarnings
filterwarnings("ignore")
```

## Step 1 - Data Ingestion

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

```python
df["species"].unique()
```

### Target feature is species
### i have 3 unique species hence it is a multiclass classification problem

## Step 2 - Perform basic data quality checks 

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
# Drop the duplicates
df = df.drop_duplicates(keep="first").reset_index(drop=True)
```

```python
df.shape
```

### Step 3 - Seperate X and Y(species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4 - Apply Preprocessing on X

```python
X.dtypes
```

### From above all values are numerical in nature
### hence it will have only numerical pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
num_pipe
```

### Step 5 - Apply train test split
test size= 0.33

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 - Build the model
Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
xtrain.columns
```

### Step 7 - Evaluate the model on test data

```python
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ytrain.head()
```

```python
ypred_train[0:5]
```

```python
ytest.head()
```

```python
ypred_test[0:5]
```

```python
yprob_test = model.predict_proba(xtest)
yprob_test.round(4)[0:5]
```

```python
model.classes_
```

### Plot the confusion matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

### Print the classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

### Crossvalidion to evaluate model properly

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring="f1_macro")
scores
```

```python
scores.mean()
```

### From above it is clear that model is good becuase most of the scores are above 0.8

# Step 8 - Out of Sample Prediction
