### Step 3 - Seperate X and Y(Weight)

1. Weight ~ Remaining features
2. id is statistically insignificant

```python
X = df.drop(columns = ["id", "Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 - Apply Preprocessing on X

```python
X.dtypes
```

```python
X.columns
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline # To combine multiple steps
from sklearn.impute import SimpleImputer # To replace missing values
from sklearn.preprocessing import StandardScaler, OneHotEncoder # For scaling num data and ohe categorical
from sklearn.compose import ColumnTransformer # To combine numerical and categorical pipeline
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

### Combine the pipelines with column transformer

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre
```

```python
pre
```

### Step 5 - Train Test Split
20% - 30% Test data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

### Step 6 - Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

### Step 7 - Model Evaluation

```python
# R2 score on train
model.score(xtrain, ytrain)
```

```python
# R2 score on test
model.score(xtest, ytest)
```

```python
ytest.head()
```

```python
from sklearn.metrics import (
    root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Training Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Testing Results : ")
evaluate_model(model, xtest, ytest)
```

### From above metrics i can conclude that this model is good as R2 > 0.8 both in train and test
1. Overfitting - Train R2 = 0.98, Test R2 - 0.65 (R2 score good in train but poor on test data)
2. Underfitting - Train R2 = 0.73, Test R2 - 0.68 (R2 score is poor both in train and test)
3. Generalized Fit - Train R2 = 0.88, Test R2 - 0.85 (R2 score > 0.8 (good) both in train and test)
